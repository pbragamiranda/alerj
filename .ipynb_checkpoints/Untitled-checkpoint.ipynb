```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

fp = './data/expenditures_transpose_30082020.csv'

df = pd.read_csv(fp, sep='|')
```

```python
df['congressman_id'] = df['congressman_id'].astype(float).astype(int) 
```

```python
df.drop_duplicates(inplace=True)
```

```python
 df.dropna(thresh=7, inplace=True)
```

```python
df.to_csv('expenditures_clean_30082020.csv', sep='|', index=False)
```
